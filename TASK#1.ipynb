```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#reading and displaying the data
url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(url);
print(df)
```

```python
df.plot(x='Hours',y='Scores',style='*',color='red')
plt.title("Hours vs Scores")
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
x=df.loc[:,['Hours']]
y=df.loc[:,['Scores']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
fun=LinearRegression();
fun.fit(x_train,y_train)
print(fun.score(x_test,y_test))
print(fun.score(x_train,y_train))
```

```python
pred=fun.predict(x_test)
print(mean_squared_error(pred,y_test))
print(np.sqrt(mean_squared_error(pred,y_test)))
```

```python
#regression line
rl=fun.coef_*x+fun.intercept_
plt.scatter(x,y)
plt.plot(x,rl,color='purple')
plt.show()
```

```python
df2=pd.DataFrame(y_test)
df2['Predicted score']=pred
print(df2)
```
